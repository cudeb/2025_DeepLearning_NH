```
import yfinance as yf
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

# âœ… 1. ë°ì´í„° ë¡œë”© ë° ì „ì²˜ë¦¬
ticker = "AAPL"          # Apple (ë¯¸êµ­)
df = yf.download(ticker, start="2022-01-01", end="2024-01-01").dropna()
```

### ë¶ˆëŸ¬ì˜¨ ì°¨íŠ¸ê°€ ì–´ë–»ê²Œ ìƒê²¼ëŠ”ì§€ í™•ì¸í•´ ë³¼ ìˆ˜ ìˆìŠµë‹ˆë‹¤!

### ğŸ“Œ ì–´ë–¤ ì§€í‘œë“¤ì„ ê°€ì§€ê³  ì£¼ê°€ë¥¼ ì˜ˆì¸¡í•  ìˆ˜ ìˆì„ê¹Œìš”?
ì£¼ê°€ì— ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ìš”ì¸ì€ ë§¤ìš° ë‹¤ì–‘í•©ë‹ˆë‹¤.
ì‹œì¥ ì „ì²´ì˜ ë¶„ìœ„ê¸°, ê²½ì œ ì§€í‘œ, ê´€ë ¨ ë‰´ìŠ¤, CEOì˜ ë°œì–¸ì´ë‚˜ ì„±í–¥ ë“± ì™¸ë¶€ì  ìš”ì¸ì´ í° ì˜í–¥ì„ ì¤„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

ê·¸ëŸ¬ë‚˜ ì´ëŸ¬í•œ ë¹„ì •í˜• ë°ì´í„°ëŠ” í˜„ì¬ ì‚¬ìš© ì¤‘ì¸ ë°ì´í„°ì…‹ì— í¬í•¨ë˜ì–´ ìˆì§€ ì•Šìœ¼ë©°, ëª¨ë“  ìš”ì†Œë¥¼ ë°˜ì˜í•˜ë ¤ë©´ ë§¤ìš° ë³µì¡í•˜ê³  ëŒ€ê·œëª¨ì˜ ëª¨ë¸ì´ í•„ìš”í•©ë‹ˆë‹¤.

ì´ë²ˆ ì‹¤ìŠµì—ì„œëŠ” ì •í˜• ë°ì´í„° ê¸°ë°˜ì˜ ë‹¨ìˆœí•œ ì˜ˆì¸¡ ëª¨ë¸ì„ êµ¬ì„±í•˜ê¸° ìœ„í•´ ë‹¤ìŒ ë‘ ê°€ì§€ ì§€í‘œë¥¼ í™œìš©í•©ë‹ˆë‹¤:

**ì „ì¼ ëŒ€ë¹„ ìˆ˜ìµë¥  (Return)**: ì˜¤ëŠ˜ ì¢…ê°€ê°€ ì „ë‚  ì¢…ê°€ì— ë¹„í•´ ì–¼ë§ˆë‚˜ ë³€í–ˆëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ë¹„ìœ¨

**20ì¼ ì´ë™í‰ê·  (20-day Moving Average)**: í˜„ì¬ ì‹œì ì„ ê¸°ì¤€ìœ¼ë¡œ, ê³¼ê±° 20ì¼ê°„ì˜ ì¢…ê°€

**20ì¼ ë³€ë™ì„± (20-day Volatility)** :  í˜„ì¬ ì‹œì ì„ ê¸°ì¤€ìœ¼ë¡œ, ê³¼ê±° 20ì¼ê°„ì˜ ì¢…ê°€ì˜ í‘œì¤€í¸ì°¨

ì´ ë‘ ì§€í‘œë¥¼ í†µí•´ ì¼ì • ìˆ˜ì¤€ì˜ íŒ¨í„´ì„ ì¶”ì¶œí•˜ì—¬ ì£¼ê°€ ì˜ˆì¸¡ì„ ì‹œë„í•´ë³´ê² ìŠµë‹ˆë‹¤.



```
# âœ… 2. ê¸°ìˆ ì  ì§€í‘œ ìƒì„±
df["Return_1"] = df["Close"].pct_change()                      # ì „ì¼ ëŒ€ë¹„ ìˆ˜ìµë¥ 
df["MA_20"] = df["Close"].rolling(20).mean()                  # 20ì¼ ì´ë™í‰ê· 
df["Volatility_20"] = df["Close"].rolling(20).std()           # 20ì¼ í‘œì¤€í¸ì°¨
```

### ë¬´ì—‡ì„ ì˜ˆì¸¡í• ê¹Œìš”?

ê²°ê³¼ê°’ì€ ìš°ë¦¬ê°€ ì›í•˜ëŠ” ëŒ€ë¡œ ê²°ì •í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

1. ë‹¤ìŒë‚ ì˜ Price( ì¢…ê°€ / ì‹œê°€ / ê³ ê°€ / ì €ê°€ ë“±...)

2. ì˜¤ëŠ˜ ëŒ€ë¹„ ë‚´ì¼ì˜ ìƒìŠ¹ë¥  (3% , -5% ë“±...)

3. ìƒìŠ¹ ì—¬ë¶€ ( ì´ì§„ ë¶„ë¥˜ - ì˜¤ë¥¸ë‹¤(1) , ë‚´ë ¤ê°„ë‹¤(0) )

ë³¸ ì‹¤ìŠµì—ì„œëŠ” **ë‹¤ìŒë‚  ì¢…ê°€ - ì˜¤ëŠ˜ ì¢…ê°€** ê°’ì„ ì˜ˆì¸¡í•˜ë„ë¡ í•˜ê² ìŠµë‹ˆë‹¤

```
# âœ… 3. íƒ€ê²Ÿ: ë‹¤ìŒë‚  ì¢…ê°€ - ì˜¤ëŠ˜ ì¢…ê°€ (íšŒê·€ìš© ì¦ê°€ëŸ‰)
df["Target"] = df["Close"].pct_change().shift(-1)
df = df.dropna()
```

```
df
```

### ì…ë ¥ ë³€ìˆ˜(X)ì™€ ì •ë‹µê°’(y) ìƒì„±
ì´ ë‹¨ê³„ì—ì„œëŠ” ëª¨ë¸ì— ì…ë ¥í•  íŠ¹ì§•(feature)ê³¼ ì˜ˆì¸¡í•˜ê³ ì í•˜ëŠ” ì •ë‹µê°’(target)ì„ ì •ì˜í•˜ê³ , ì •ê·œí™”ë¥¼ ìˆ˜í–‰í•©ë‹ˆë‹¤.

**ì…ë ¥ ë³€ìˆ˜(X)**

Return_1: ì „ì¼ ëŒ€ë¹„ ìˆ˜ìµë¥ 

MA_20: 20ì¼ ì´ë™í‰ê· 

Volatility_20: 20ì¼ê°„ì˜ í‘œì¤€í¸ì°¨ (ë³€ë™ì„±)

**ì •ë‹µê°’(y)**

Target: ë‹¤ìŒë‚ ì˜ ìˆ˜ìµë¥  (ì˜ˆì¸¡ ëª©í‘œ)

ì´í›„ ëª¨ë¸ í•™ìŠµì„ ìœ„í•´ StandardScalerë¥¼ ì‚¬ìš©í•˜ì—¬ ì…ë ¥(X)ê³¼ ì •ë‹µ(y)ì„ í‰ê·  0, í‘œì¤€í¸ì°¨ 1ì˜ ì •ê·œë¶„í¬ë¡œ ì •ê·œí™”í•©ë‹ˆë‹¤.
ì´ ì •ê·œí™” ê³¼ì •ì€ ë”¥ëŸ¬ë‹ ëª¨ë¸ì˜ ìˆ˜ë ´ ì†ë„ë¥¼ ë†’ì´ê³ , ì„±ëŠ¥ì„ ì•ˆì •í™”ì‹œí‚¤ëŠ” ë° ë„ì›€ì„ ì¤ë‹ˆë‹¤.

```
# âœ… 5. ì…ë ¥(X)ì™€ ì •ë‹µ(y) ìƒì„±
features = ["Return_1", "MA_20", "Volatility_20"]
X = df[features].values
y = df["Target"].values.reshape(-1, 1)

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)
```

### í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„ë¦¬ ë° í…ì„œ ë³€í™˜
**í•™ìŠµìš© / í…ŒìŠ¤íŠ¸ìš© ë°ì´í„° ë¶„ë¦¬**

ì „ì²´ ë°ì´í„°ì˜ 80%ë¥¼ í•™ìŠµ(train), 20%ë¥¼ í…ŒìŠ¤íŠ¸(test) ë°ì´í„°ë¡œ ì‚¬ìš©í•©ë‹ˆë‹¤.

**PyTorch í…ì„œë¡œ ë³€í™˜**

ë”¥ëŸ¬ë‹ ëª¨ë¸ì—ì„œ í•™ìŠµí•  ìˆ˜ ìˆë„ë¡ NumPy ë°°ì—´ì„ PyTorch í…ì„œë¡œ ë³€í™˜í•©ë‹ˆë‹¤.


ì…ë ¥(X)ì€ ì •ê·œí™”ëœ ê°’ì„ ì‚¬ìš©í•˜ê³ ,

ì •ë‹µ(y)ì€ ì•ì„œ ì •ê·œí™”ëœ y_scaled ê°’ì„ ê¸°ë°˜ìœ¼ë¡œ ë¶„í• ëœ ì‹œì ì— ë§ê²Œ ë‚˜ëˆ ì¤ë‹ˆë‹¤.


nhì˜ˆì‹œê·¸ë¦¼.png





```
# âœ… 7. train/test ë¶„ë¦¬
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, shuffle=False)
```

```
# âœ… 8. í…ì„œ ë³€í™˜
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_train = torch.tensor(y_scaled[:len(X_train)], dtype=torch.float32)
y_test = torch.tensor(y_scaled[len(X_train):], dtype=torch.float32)
```

## 9. PyTorch ê¸°ë°˜ íšŒê·€ ëª¨ë¸ ì •ì˜
ì´ì œ PyTorchë¡œ ê°„ë‹¨í•œ ë‹¤ì¸µ í¼ì…‰íŠ¸ë¡ (Multi-Layer Perceptron, MLP) êµ¬ì¡°ì˜ íšŒê·€ ëª¨ë¸ì„ ì •ì˜í•©ë‹ˆë‹¤.

**ì…ë ¥ê°’**ì€ 3ê°œ ì§€í‘œ(Return_1, MA_20, Volatility_20), **ì¶œë ¥**ì€ ë‹¤ìŒë‚  ìˆ˜ìµë¥ (ì—°ì†í˜• ì‹¤ìˆ˜) í•˜ë‚˜ì…ë‹ˆë‹¤.

###**ì£¼ìš” ì½”ë“œ ì„¤ëª…**

nn.Linear(input_dim, 32)
â†’ 3ì°¨ì› ì…ë ¥(3ê°œ ì§€í‘œ)ì„ ë°›ì•„ 32ì°¨ì›ìœ¼ë¡œ ë³€í™˜ (ì²« ë²ˆì§¸ ì€ë‹‰ì¸µ)

nn.ReLU()
â†’ ë¹„ì„ í˜•ì„±ì„ ì£¼ê¸° ìœ„í•œ ReLU í•¨ìˆ˜ ì‚¬ìš©

nn.Linear(16, 1)
â†’ ë§ˆì§€ë§‰ ì¶œë ¥ì€ ì˜ˆì¸¡í•  ìˆ˜ìµë¥  í•˜ë‚˜, ë”°ë¼ì„œ ì¶œë ¥ ë…¸ë“œëŠ” 1ê°œ

forward() í•¨ìˆ˜
â†’ í•™ìŠµ/ì˜ˆì¸¡ ì‹œ ì‹¤ì œ ë°ì´í„°ê°€ í†µê³¼í•˜ëŠ” ê²½ë¡œë¥¼ ì •ì˜

model = StockRegressor(input_dim=3)
-> ìš°ë¦¬ëŠ” 3ê°œì˜ inputê°’ (3ê°œì˜ ì§€í‘œ)ë¥¼ ì‚¬ìš©í•  ê²ƒì´ê¸°ì—, input_dimì€ 3 ì…ë‹ˆë‹¤

ê·¸ë¦¼3.png


```
# âœ… 9. íšŒê·€ ëª¨ë¸ ì •ì˜
class StockRegressor(nn.Module):
    def __init__(self, input_dim):
        super(StockRegressor, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1)  # íšŒê·€ëŠ” ì¶œë ¥ì´ 1ê°œ
        )

    def forward(self, x):
        return self.model(x)

model = StockRegressor(input_dim=3)
```

###í•™ìŠµ ì„¤ì •
ëª¨ë¸ì„ í•™ìŠµì‹œí‚¤ê¸° ìœ„í•´ ì†ì‹¤ í•¨ìˆ˜ì™€ ìµœì í™” ì•Œê³ ë¦¬ì¦˜ì„ ì„¤ì •í•©ë‹ˆë‹¤.

**criterion** -> Regressionì—ì„œ ì£¼ë¡œ ì‚¬ìš©ë˜ëŠ” loss í•¨ìˆ˜

**epochs** -> í•™ìŠµì´ ë°˜ë³µë  íšŸìˆ˜

```
# âœ… 10. í•™ìŠµ ì„¤ì •
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.005)
epochs = 500
```

```
# âœ… 11. í•™ìŠµ ë£¨í”„
for epoch in range(epochs):
    model.train()
    preds = model(X_train)
    loss = criterion(preds, y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}] | Loss: {loss.item():.6f}")
```

### ëª¨ë¸ í‰ê°€ (ì˜ˆì¸¡ ë° ì„±ëŠ¥ ì§€í‘œ ê³„ì‚°)
ëª¨ë¸ í•™ìŠµì´ ì™„ë£Œë˜ë©´, í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¥¼ í™œìš©í•˜ì—¬ ì˜ˆì¸¡ ì„±ëŠ¥ì„ í‰ê°€í•©ë‹ˆë‹¤.

```
# âœ… 5. í‰ê°€
model.eval()
y_pred_scaled = model(X_test).detach().numpy()
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_true = scaler_y.inverse_transform(y_test.numpy())
with torch.no_grad():
    y_pred = model(X_test).numpy()
    y_true = y_test.numpy()

    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)


    print(f"\nğŸ“Š MSE: {mse:.4f}")
    print(f"ğŸ“Š MAE: {mae:.4f}")

```

```
# âœ… 13. ì˜ˆì¸¡ vs ì‹¤ì œ ì¦ê°€ëŸ‰ ì‹œê°í™”
plt.figure(figsize=(10, 4))
plt.plot(y_true, label="True Change", linewidth=2)
plt.plot(y_pred, label="Predicted Change", linestyle='--')
plt.title("Stock Price Change Prediction")
plt.xlabel("Time (Test Set)")
plt.ylabel("Price Change (KRW)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
