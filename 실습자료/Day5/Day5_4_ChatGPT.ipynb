  - 에러 처리 실습
   - (아래 코드와 에러 메시지를 복사한 이후) 해당 코드를 실행했을 때 다음과 같은 에러가 발생했는데, 에러의 원인을 알려주시고 이를 해결할 수 있는 새로운 수정된 코드도 함께 주세요.

```python
import torch
import torch.nn as nn

class MyNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 128)
        self.fc2 = nn.Linear(128, 64)
        self.fc3 = nn.Linear(64, 10)

    def forward(self, x):
        # x shape: (batch_size, 28, 28)
        x = x.view(x.size(0), -1)  # flatten to (batch_size, 784)
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        return x

model = MyNet()

# 예를 들어 배치 크기 3, 흑백 이미지 28x28
image = torch.randn(3, 28, 28)  # batch_size=3

output = model(image)
print(output.shape)  # torch.Size([3, 10])
```
