#### 1.1 Tensor 생성

- Data로부터 Tensor 생성
  - `import torch` : PyTorch를 사용하기 위한 라이브러리를 호출
  - `torch.tensor(data)` : **Data를 Tensor로 변환**
  - `Tensor.shape` : Tensor의 크기를 반환

```python
import torch

data_1d = [1,2,3,4,5] # 1차원 데이터
tensor_1d = torch.tensor(data_1d) # 1차원 Tensor
print(f"데이터 : {data_1d}, \nTensor : {tensor_1d}")
print(f"Tensor의 크기 : {tensor_1d.shape}")
```

```python
data_2d = [[1,2],[2,3],[3,4]] # 2차원 데이터
tensor_2d = torch.tensor(data_2d) # 2차원 Tensor
print(f"데이터 : {data_2d}, \nTensor : {tensor_2d}")
print(f"Tensor의 크기 : {tensor_2d.shape}")
```

```python
data_3d = [[[1,2,3,4],[2,3,4,5],[3,4,5,6]],[[7,8,9,10],[10,11,12,13],[14,15,16,17]]] # 3차원 데이터
tensor_3d = torch.tensor(data_3d) # 3차원 Tensor
print(f"데이터 : {data_3d}, \nTensor : {tensor_3d}")
print(f"Tensor의 크기 : {tensor_3d.shape}")
```

- Random한 값으로 Tensor를 생성

  - `torch.rand(shape)` : 랜덤한 값으로 초기화된 Tensor를 생성

  - `torch.ones(shape)` : 1로 이루어진 Tensor를 생성

  - `torch.zeros(shape)` : 0으로 이루어진 Tensor를 생성

```python
shape = (2,3)

tensor_rand = torch.rand(shape)
print(f"[torch.rand] \n{tensor_rand}\n")

tensor_ones = torch.ones(shape)
print(f"[torch.ones] \n{tensor_ones}\n")

tensor_zeros = torch.zeros(shape)
print(f"[torch.zeros] \n{tensor_zeros}")
```

#### 1.2 Tensor의 데이터 타입

- Tensor는 다양한 데이터 타입을 가질 수 있음
  - 정수형
    - `torch.int8`, `torch.int16`, `torch.int32`, `torch.int64`(=`torch.long`)
    - `torch.uint8`
  - 실수형
    - `torch.float16`, `torch.float32`, `torch.float64`
  - Boolean형
    - `torch.bool`
  - ETC

- `Tensor.dtype`을 통해 Tensor의 데이터 타입을 확인가능

```python
data_int = [[1,2],[2,3]]
tensor_int = torch.tensor(data_int)
print(f"[tensor_int] \n{tensor_int}")
print(f"데이터 타입 : {tensor_int.dtype}\n")

data_float = [[5.3,8.9],[3.5,7.3]]
tensor_float = torch.tensor(data_float)
print(f"[tensor_float] \n{tensor_float}")
print(f"데이터 타입 : {tensor_float.dtype}\n")

data_boolean = [[True,False],[False,False]]
tensor_boolean = torch.tensor(data_boolean)
print(f"[tensor_boolean] \n{tensor_boolean}")
print(f"데이터 타입 : {tensor_boolean.dtype}\n")
```

- Tensor의 데이터 타입 지정 및 변경
  - `torch.tensor`의 `dtype` 파라미터에 데이터 타입을 넘겨줌으로써, 원하는 데이터 타입을 지정하여 Tensor를 생성할 수 있음
  - `.to(dtype)` 을 통해 tensor의 데이터 타입을 변경할 수 있음

```python
# 따로 타입을 지정하지 않는 경우 기존 데이터의 타입을 따라서 정수형 Tensor가 생성됨
data_int = [[1,2],[2,3]]
tensor_int = torch.tensor(data_int)
print(f"[tensor_int] \n{tensor_int}")
print(f"데이터 타입 : {tensor_int.dtype}\n")

# 타입을 실수형(torch.float32)으로 지정하면, 실수형 Tensor가 생성됨
data_int = [[1,2],[2,3]]
tensor_float = torch.tensor(data_int,dtype=torch.float32)
print(f"[tensor_float] \n{tensor_float}")
print(f"데이터 타입 : {tensor_float.dtype}\n")
```

```python
# 정수형 Tensor 생성
data_int = [[1,2],[2,3]]
tensor_int = torch.tensor(data_int)
print(f"[tensor_int] \n{tensor_int}")
print(f"데이터 타입 : {tensor_int.dtype}\n")

# 실수형 Tensor로 변경
tensor_float = tensor_int.to(torch.float32)
print(f"[tensor_float] \n{tensor_float}")
print(f"데이터 타입 : {tensor_float.dtype}")
```

#### 1.3 Tensor의 Device

- PyTorch에서는 **GPU**를 활용한 연산을 지원합니다.

- Tensor의 Device를 GPU로 설정하여, 이 Tensor로 연산을 진행할 때 GPU를 사용하겠다는 것을 명시

- 기본적으로 Tensor의 Device는 CPU

- `Tensor.device`를 통해서 현재 Tensor의 device를 확인할 수 있음

```python
data_int = [[1,2],[2,3]]
tensor = torch.tensor(data_int)
print(f"Tensor의 Device : {tensor.device}")
```

- `torch.cuda.is_available()` 를 통해서 현재 GPU를 사용할 수 있는 환경인지 확인할 수 있음
- `torch.device` 를 통해 device를 명시할 수 있음
- `Tensor.to(device)` 를 통해서 Tensor의 device를 설정할 수 있음

```python
if torch.cuda.is_available():
  print("GPU 사용 가능!")
  device = torch.device("cuda")
  tensor = tensor.to(device)
else:
  print("GPU 사용 불가능!")
  device = tensor.to("cpu")
  tensor = tensor.to(device)

print(f"Tensor의 Device : {tensor.device}")
```

```python
# 위의 코드를 줄인 버전
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
tensor = tensor.to(device)
print(f"Tensor의 Device : {tensor.device}")
```

#### 1.4 Tensor의 연산

- Indexing & Slicing


```python
# [3,3] 크기의 Tensor 생성
data = [[1,2,3],[4,5,6],[7,8,9]]
tensor = torch.tensor(data)
```

```python
# 행 Indexing
print(f"첫번째 행: {tensor[0]}")
print(f"마지막 행: {tensor[-1]}\n") # tensor[2]

# 열 Indexing
print(f"첫번째 열: {tensor[:, 0]}")
print(f"마지막 열: {tensor[:, -1]}\n") # tensor[:,2]

# 요소 Indexing
print(f"첫번째 요소: {tensor[0,0]}")
print(f"가운데 요소: {tensor[1,1]}")
print(f"마지막 요소: {tensor[-1,-1]}") # tensor[2,2]
```

```python
# 행 Slicing
print(f"첫번째, 두번째 행: \n{tensor[0:2]}\n")

# 열 Slicing
print(f"첫번째, 두번째 열: \n{tensor[:,0:2]}\n")

# 행 & 열 Slicing
print(f"첫번째, 두번째 행과 첫번째, 두번째 열: \n{tensor[0:2,0:2]}")
```

- Concatenation

- 사칙연산 & 행렬 곱 & 행렬 Transpose
  - 두 Tensor의 사칙 연산은 element-wise 하게 연산
  - 같은 크기의 Tensor 사이에만 사칙 연산이 가능
    - `add(+)`, `sub(-)`, `mul(*)`, `div(/)`
  - 두 Tensor의 행렬 곱 : `matmul(@)`
  - Tensor의 행렬 Transpose : `Tensor.T`

```python
# 1로 이루어진 Tensor 2개 생성
tensor1 = torch.ones(2,2)
tensor2 = torch.ones(2,2)
```

```python
# 덧셈
tensor_add = tensor1 + tensor2 # tensor_add = tensor1.add(tensor2)
print(f"[tensor_add] : \n{tensor_add}\n")

# 뺄셈
tensor_sub= tensor1 - tensor2 # tensor_sub = tensor1.sub(tensor2)
print(f"[tensor_sub] : \n{tensor_add}\n")

# 곱셈
tensor_mul= tensor1 * tensor2 # tensor_mul = tensor1.mul(tensor2)
print(f"[tensor_mul] : \n{tensor_mul}\n")

# 나눗셈
tensor_div= tensor1 / tensor2 # tensor_div = tensor1.div(tensor2)
print(f"[tensor_div] : \n{tensor_div}\n")
```

```python
# Transpose
tensor = torch.ones(2,3)
tensor_transpose = tensor.T
print(f"[tensor] : \n{tensor}\n")
print(f"[tensor_transpose] : \n{tensor_transpose}")
```

```python
# 행렬 곱
tensor1 = torch.ones(2,3)
tensor2 = torch.ones(3,2)

tensor_matmul = tensor1 @ tensor2

print(f"[tensor1] : \n{tensor1}\n")
print(f"[tensor2] : \n{tensor2}\n")
print(f"[tensor_matmul] : \n{tensor_matmul}")
```
